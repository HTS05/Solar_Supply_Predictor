```python
import pandas as pd
df = pd.read_csv("GUWAHATI IRRIDANCE.csv", skiprows=18)
df.head()
```

```python
print(df.isnull().sum())
# check if any row has at least one null value
print(df.isnull().any(axis=1).sum())
```

```python
# Check for negative values in all numerical columns
negative_values = (df.select_dtypes(include='number') < 0)
print(negative_values.sum())

# To display rows with any negative value in numeric columns
rows_with_negatives = df[negative_values.any(axis=1)]
print(rows_with_negatives)
```

```python
# Combine YEAR, MO, and DY columns into a single datetime column named 'Date'
df = df.rename(columns={'YEAR': 'year', 'MO': 'month', 'DY': 'day'})
df['Date'] = pd.to_datetime(df[['year', 'month', 'day']])
df.head()


# Now you can extract month and day if needed
# df['Month'] = df['Date'].dt.month
# df['Day'] = df['Date'].dt.day

# Optionally, drop the original YEAR, MO, and DY columns if you don't need them anymore
# df.drop(columns=['YEAR', 'MO', 'DY'], inplace=True)
```

```python
df.rename(columns={
    'ALLSKY_SFC_SW_DWN': 'GHI',
    'T2M': 'Temperature',
    'RH2M': 'Humidity',
    'CLOUD_AMT': 'Cloud_Coverage',
    'WS2M': 'Wind_Speed',
    'PRECTOTCORR': 'Precipitation',
    'PS': 'Pressure',
    'WSC': 'Corrected_Wind_Speed'
}, inplace=True)
df.head()
```

```python
# Drop the extra year, month, day columns
df = df.drop(columns=['year', 'month', 'day', 'Month', 'Day'], errors='ignore')

# Move 'Date' to first column
columns = df.columns.tolist()
columns.insert(0, columns.pop(columns.index('Date')))
df = df[columns]

df.head()
```

```python
df.to_csv("GUWAHATI_IRRIDANCE_cleaned.csv", index=False)
```
